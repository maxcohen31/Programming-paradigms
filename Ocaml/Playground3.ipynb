```ocaml
(* Luglio 2023 *)

(*

    bimedia prende una lista di coppie di interi lis che si assume non vuota e restituisce la coppia (m1,m2)
    dove m1 è la media dei primi elementi delle coppie in lis ed m2 è la media dei secondi elementi delle
    stesse coppie. 
    Ad esempio:
    
    bimedia [(1,20);(5,10);(6,30)]  restituisce (4,20) 
    
    alternati prende una lista di coppie di interi lis e restituisce la somma dei primi elementi delle coppie
    in posizioni dispari nella lista con i secondi elementi delle coppie in posizioni pari. In altre parole, se
    lis = [(a1,b1);(a2,b2);(a3,b3);(a4,b4);...] calcola a1+b2+a3+b4+.... Ad esempio:
    
    alternati [(1,10);(2,20);(3,30);(4,40)] restituisce 64

*)

let bimedia lis =
    let rec aux l acc1 acc2  = 
        match l with
        | [] -> (acc1, acc2)
        | x :: [] -> (fst x + acc1 , snd x + acc2)
        | x :: y :: tl -> aux tl (acc1 + fst x + fst y) (acc2 + snd x + snd y)
        
    in let (s1, s2) = aux lis 0 0 
    in (s1 / List.length lis, s2 / List.length lis) ;;
    
    
let rec alternati lis = 
    match lis with
    | [] -> 0
    | (x, y) :: [] -> 0
    | (x, y) :: (z, w) :: tl -> x + w + alternati tl ;; 
    
alternati [(1,10);(2,20);(3,30);(4,40)] ;;
```

```ocaml
(* Settembre 2023 *)

type colored_tree =
    | BlackNode of int * colored_tree * colored_tree
    | RedNode of int * colored_tree * colored_tree
    | GreenLeaf of int ;;

let ct = BlackNode (10,RedNode (5,BlackNode (2,GreenLeaf 1,GreenLeaf 2),GreenLeaf 4),RedNode (7,GreenLeaf 2,GreenLeaf 5)) ;;
    
(*

    color_tree_lists ct restituisca una tripla i cui tre elementi sono liste contenenti tutti i valori interi
    presenti rispettivamente in nodi neri, rossi e verdi dell’albero;
    
    color_tree_lists ct restituisce ([10; 2], [5; 7], [1; 2; 4; 2; 5]).

*)

let color_tree_lists ct = 
    let rec aux tree acc1 acc2 acc3 = 
        match tree with 
        | GreenLeaf n -> (acc1, acc2, n::acc3)
        | RedNode (n, left, right) -> let (l1, l2, l3) = aux left acc1 acc2 acc3 in
                                     let (r1, r2, r3) = aux right acc1 acc2 acc3 in
                                     (l1 @ r1, n :: l2 @ r2, l3 @ r3)
                                     
        | BlackNode (n, left, right) -> let (l1, l2, l3) = aux left acc1 acc2 acc3 in
                                       let (r1, r2, r3) = aux right acc1 acc2 acc3 in
                                       (n :: l1 @ r1, l2 @ r2, l3 @ r3)
    in aux ct [] [] [] ;;
    
(*
    color_forest_lists ctlis esegue lo stesso calcolo di color_tree_lists, ma su una lista di alberi
    colorati, restituendo una unica tripla i cui elementi contengono tutti i valori presenti rispettivamente
    in nodi neri, rossi e verdi di tutti gli alberi della lista. (Nota: color_forest_lists può usare al suo
    interno color_tree_lists. )
*)

let ctlis = [BlackNode (1,GreenLeaf 2,GreenLeaf 3);RedNode (4,GreenLeaf 5,GreenLeaf 6);BlackNode (6,RedNode (7,GreenLeaf 8,GreenLeaf 9),RedNode (10,GreenLeaf 11,GreenLeaf 12))]

(*
    color_forest_lists ctlis restituisce ([1; 6], [4; 7; 10], [2; 3; 5; 6; 8; 9; 11; 12])
*)

let color_forest_lists ctlis =  
    match ctlis with
    | [] -> ([], [], [])
    | tl -> let list_of_colored_nodes = List.map color_tree_lists tl
            in let merge (b1, r1, g1) (b2, r2, g2) = (b1 @ b2, r1 @ r2 , g1 @ g2)
            in List.fold_left merge ([], [], []) list_of_colored_nodes ;;
    
    
color_forest_lists ctlis ;;
```

```ocaml
(* Esercizio simulazione d'esame *)

type ntree = 
    | Node of int * ntree list
    | Leaf of int ;;
    
(*
    Definire, usando i costrutti di programmazione funzionale in OCaml, una funzione compact con tipo
    
    compact : ntree -> ntree

    tale che compact t restituisca un albero t’ ottenuto sostituendo ogni nodo Node (n,[Leaf n1;...;Leaf nk])
    (ossia che non ha figli di tipo Node) contenuto in t, con un nodo Node (n,[Leaf (n1+...+nk)])
*)


let rec compact tr = 
    (* funzione che controlla se il parametro passato è una folgia o no*)
    let isLeaf l = 
        match l with
        | Leaf x -> true
        | Node (n, c) -> false
    in
        (* funzione che somma i valori delle foglie in una lista facendo pattern matching sul tipo Leaf *)
       let sum_leaves lis = List.fold_left (fun acc -> function Leaf x -> acc + x | _ -> acc) 0 lis
    in 
       match tr with
       | Leaf x -> Leaf x
       | Node (n, children) -> (* se i figli sono tutte foglie faccio la somma *)
                               if (List.for_all isLeaf children) 
                               then Node (n, [Leaf (sum_leaves children)])
                               (* altrimenti chiamo la funzione sul sottoalbero *)
                               else Node (n, List.map compact children) ;;
            
            
            
let t1 = Node(10, [Leaf 2; Node (3, [Leaf 5; Leaf 3]); Leaf 4]) ;;
compact t1 ;;
```

```ocaml
(* esempio di esame 

    Definire, usando i costrutti di programmazione funzionale in OCaml, una funzione duemax con tipo

    duemax : ‘a list -> (‘a * ‘a) option
    
    in modo che duemax lis restituisca Some (x,y) tale che per ogni elemento z di lis vale x ≥ y ≥ z. La
    funzione restituisce None in caso di lista vuota. Esempi:
    
    duemax [2;3;1;5] = Some (5,3)
    duemax [’t’] = Some (’t’,’t’)
    duemax [] = None
*)

let rec duemax lis = 
    match lis with
    | [] -> None
    | x :: y :: [] -> if x < y then Some(y, x) else Some(x, y)
    | z :: tl -> match duemax tl with
                 | None -> failwith "Error"
                 | Some (a, b) -> if z >= a then Some (z, a)
                                  else if z >= b then Some (a, z)
                                  else Some (a, b) ;;




duemax [2; 3; 1; 5] ;;
```

```ocaml
(*
    Senza utilizzare esplicitamente la ricorsione, ma utilizzando funzioni higher-order su liste (dal modulo List),
    si definisca in OCaml una funzione split con tipo
    
    split: ‘a list -> (‘a -> bool) -> (‘a list * ‘a list)
    
    in modo che split lis p restituisca una copia (lis1,lis2) tale che la lista lis1 contenga tutti e soli gli
    elementi di lis che soddisfano il predicato p e la lista lis2 contenga
    tutti e soli gli elementi di lis che non soddisfano il predicato p. 
    La soluzione dovrebbe fare in modo che la lista lis venga scandita una sola volta.

*)

let split lis p =
    let rec aux l acc1 acc2 = 
        match l with
        | [] -> ([], [])
        | x :: [] -> if p x then (x::acc1, acc2) else (acc1, x::acc2)
        | x :: tl -> if p x then aux tl (x::acc1) acc2 else aux tl acc1 (x :: acc2) 
    in aux lis [] [] ;;
    
(* higher-order function *)
let split2 lis p = 
    match lis with
    | [] -> ([], [])
    | l -> (List.filter p l, List.filter (fun x -> not (p x)) l) ;;

(* soluzione proposta *)
let split3 lis p = 
    let f (lis1, lis2) x = if p x then (x::lis1, lis2) else (lis1, x::lis2) in List.fold_left f ([], []) lis ;;

split2 [1; 2; 3; 4] (fun x -> x mod 2 = 0) ;;
```

```ocaml
(* 
    Giugno 2022 
    Si definisca, usando i costrutti di programmazione funzionale di OCaml, una funzione merge con tipo
    merge : ’a list list -> ’a list

    tale che, data una lista di liste lis, l’applicazione merge lis restituisca una lista contenente tutti gli
    elementi delle liste contenute in lis, senza ripetizioni.
    Ad esempio, 
    merge [[3;4;5];[9;5];[3;6;9;6]] deve restituire la lista [3;4;5;9;6]. 
    L’ordine degli elementi nel risultato può essere qualunque.
    
*)

let filter_duplicates lis = 
    let rec aux l acc = 
        match l with 
        | [] -> acc
        | x :: tl -> if List.mem x acc then aux tl acc else aux tl (x :: acc) 
    in aux lis [] ;;

let flat lis = 
    let rec aux l acc = 
        match l with
        | [] -> acc
        | x :: [] -> x @ acc
        | x :: tl -> x @ acc @ aux tl acc
    in aux lis [] ;;
    
let merge lis = filter_duplicates (flat lis) ;;   

let l2 = [[3;4;5];[9;5];[3;6;9;6]] ;;
merge l2 ;;
```

```ocaml
(* Permutation of a list *)

(* remove a element from a list at a given index *)
let rec remove_at i lis = 
    match lis with
    | [] -> failwith "Error: list is empty"
    | x :: tl -> if i = 0 then tl else x :: remove_at (i-1) tl ;;
    

let remove_el lis idx = 
    let rec aux l acc n = 
        match l with
        | [] -> failwith "Error: empty list"
        | x :: tl -> if n = 0 then (x, acc @ tl) else aux tl (x :: acc) (n - 1) 
    in aux lis [] idx ;;
    
let rand_remove lis = let idx = Random.int (List.length lis) in remove_el lis idx ;;  


let permutation lis = 
    let rec aux l acc = 
        match l with
        | [] -> acc
        | rest -> let (el_extracted, rest_list) = rand_remove rest in aux rest_list (el_extracted :: acc) 
    in aux lis [] ;;
    
(* all together *)
let permutation2 lis = 
    let rec remove_el l acc idx = 
        match l with
        | [] -> failwith "Error: empty list"
        | x :: tl -> if idx = 0 then (x, acc @ tl) else remove_el tl (x :: acc) (idx - 1)
    in
    let rec rand_remove l = 
        let i = Random.int (List.length l) 
    in remove_el l [] i
    in
    let rec aux l acc = 
        match l with
        | [] -> acc
        | rest -> let (el_extracted, rest_list) = rand_remove rest in aux rest_list (el_extracted :: acc) 
    in aux lis [] ;;

let l = ["a"; "b"; "c"; "e"; "f"] ;;
remove_el l 4 ;;
rand_remove l ;; 
(* permutatation l *)
permutation2 l ;;
```

```ocaml
(* Collect the Nodes at a Given Level in a List *)
type 'a node = 
    | Empty
    | Node of 'a * 'a node * 'a node ;;  
    
    
let collect_at tr lv = 
    let rec aux t acc liv = 
        match t with
        | Empty -> acc
        | Node (x, l, r) -> if lv = liv then (x :: acc) else 
                            let left = aux l acc (liv + 1) in
                            let right = aux r acc (liv + 1) in
                            left @ right @ acc
    in aux tr [] 1 ;;

(*  
    Proposed solution 

     Having an accumulator acc prevents using inefficient List.append.
     Every node at level N will be pushed directly into accumulator.
     Not tail-recursive, but that is no problem since we have a binary tree and
     and stack depth is logarithmic. 
*)
  let at_level t level =
    let rec at_level_aux t acc counter = match t with
      | Empty -> acc
      | Node (x, l, r) ->
        if counter=level then
          x :: acc
        else
          at_level_aux l (at_level_aux r acc (counter + 1)) (counter + 1)
    in
      at_level_aux t [] 1;;
    
let example_tree =
  Node ('a', Node ('b', Node ('d', Empty, Empty), Node ('e', Empty, Empty)),
       Node ('c', Empty, Node ('f', Node ('g', Empty, Empty), Empty))) ;;
       
collect_at example_tree 2 ;;
```

```ocaml
(* Sorting a List of Lists According to Length of Sublists - List.sort is not allowed :) *)

(* Quicksort - it takes a comparing function to do the sorting *)
let partition lis cmp pivot = 
    let rec aux l acc1 acc2 = 
        match l with
        | [] -> (acc1, acc2)
        | x :: tl -> if cmp x pivot then aux tl (x::acc1) acc2 else aux tl acc1 (x::acc2)
    in aux lis [] [] ;;
    
let rec quicksort cmp lis = 
    match lis with
    | [] -> []
    | x :: list' -> let (l1, l2) = partition list' cmp x  in quicksort cmp l1 @ [x] @ quicksort cmp l2 ;;
    
let sort_by_length lis = quicksort (fun x y -> List.length x <= List.length y) lis ;;
        
(* association list  - [(len, frequency)] *)
let frequency_count lis = 
    let lengths = List.map (fun x -> List.length x) lis in
    (* function used by fold_left - freq_lis è l'accumulatore della fold_left *)
    let rec aux key freq_lis = 
        match freq_lis with
        | [] -> [(key, 1)]
        | (len, c) :: tl -> if len = key then (len, c+1) :: tl else (len, c) :: aux key tl
    in List.fold_left (fun acc x -> aux x acc) [] lengths ;; 
       
let sort_by_freq lis = 
    let freq_list = frequency_count lis in 
    
    (* helper function: given a length it returns thow many times a list with that length appears *)
    let get_frequency len = 
        let rec aux l = 
            match l with
            | [] -> 0
            | (x, y) :: tl -> if len = x then y else aux tl 
        in aux freq_list
    (* sorting the list by using the get_frequency list function *)
    in quicksort (fun x y -> get_frequency (List.length x) <= get_frequency (List.length y)) lis ;;
        
let l = [["a"; "b"; "c"]; ["d"; "e"]; ["f"; "g"; "h"]; ["d"; "e"]; ["i"; "j"; "k"; "l"]; ["m"; "n"]; ["o"]] ;;
frequency_count l ;;
sort_by_freq l ;;
```

```ocaml
(* pack consecutive duplicates *)

let pack lis = 
    let rec aux l curr acc = 
        match l with
        | [] -> []
        | [x] -> (x :: curr) :: acc 
        | x :: (y :: _ as tl) -> if x = y then aux tl (x::curr) acc else aux tl [] ((x::curr)::acc)
    in List.rev (aux lis [] []) ;;

let list = [1; 1; 1; 2; 2; 3; 1; 1; 4] ;;
pack list ;;
```

```ocaml
(* Construct a binary tree from a list *)
type 'a binary_tree = 
    | Empty
    | Node of 'a * 'a binary_tree * 'a binary_tree ;;
    

let rec insert tr v = 
    match tr with
    | Empty -> Node(v, Empty, Empty)
    | Node(x, l, r) -> if v = x then tr 
                       else if v < x then Node(x, insert l v, r)
                       else Node(x, l , insert r v) ;;

let construct lis = 
    match lis with
    | [] -> Empty
    | x :: tl -> List.fold_left (fun acc x -> insert acc x) Empty tl ;;

let l1 = [3; 2; 5; 7; 1] ;;
construct l1 ;;
```
