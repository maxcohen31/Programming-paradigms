```ocaml
(* lista palindroma *)
let reverse lis = 
   let rec aux l acc  = 
       match l with
       | [] -> acc
       | x :: tl -> aux tl (x :: acc) 
    in aux lis [] ;;
    
let is_palindrome l = l = reverse l ;;
not (is_palindrome ["x"; "a"; "m"; "a"; "x"]) ;;
is_palindrome [1; 2; 2; 1] ;;
```

```ocaml
(* 
    Si definisca una funzione OCaML a tre argomenti insert_at che prende un elemento x, 
    un indice n ed una lista lst, e inserisce x in posizione n in lst.
*)

let insert_at x n lst = 
    let rec aux l idx = 
        match l with
        | [] -> [x]
        | el :: tl -> if idx = n then x :: el :: tl else el :: aux tl (idx + 1)
    in aux lst 0 ;;
    
insert_at "alfa" 1 ["a"; "b"; "c"; "d"] ;; 
```

```ocaml
(*
    Si definisca una funzione OCaML range che prende due interi a e b 
    e restituisce una lista contenenti tutti gli interi nell'intervallo da a a b.
*)

let rec range a b = 
    match a, b with
    | x, y -> if x = y then [x] else if x < y then x :: range (x+1) y else x :: range (x-1) y ;;
    
range 9 4 ;;
range 4 9 ;; 
```

```ocaml
type tree = 
    | Empty
    | Node of tree * int * tree ;;


(* funzione che somma i valori contenuti nei nodi di un albero *)
let sum tree = 
    let rec aux tr acc = 
        match tr with
        | Empty -> acc
        | Node (l, v, r) -> let sum_left = aux l acc in
                            let sum_right = aux r acc in
                            v + sum_left + sum_right 
    in aux tree 0 ;;

(* funzione che conta il numero di nodi in un albero *)
let nodes_count tree = 
    let rec aux tr acc =
        match tr with
        | Empty -> acc
        | Node (Empty, v, Empty) -> acc + 1
        | Node (l, v, r) -> let left = aux l acc in
                            let right = aux r acc in
                            1 + left + right 
    in aux tree 0 ;;
                            

(* 
    Si definisca una funzione OCaML 
    depth che prende un albero binario t e restituisce la lunghezza del cammino massimo.
*)

let rec height tree = 
    match tree with
    | Empty -> 0
    | Node (l, v, r) -> let max a b = if a < b then b else b in
                        let left_h = height l in
                        let right_h = height r in
                        max left_h right_h + 1 ;;

(*
    Si definisca una funzione OCaML contains che prende un albero binario t ed un valore intero n 
    e restituisce true se n è contenuto in un nodo di t false altrimenti
*)
let rec contains tree x = 
    match tree with
    | Empty -> false
    | Node (l, v, r) -> if v = x then true else 
                        let left = contains l x in
                        let right = contains r x in
                        left || right ;;
                        
(*
    Si definisca una funzione OCaML equals che prende due alberi binari 
    t1 e t2 e restituisce true se t1 e t2 sono lo stesso albero false altrimenti
*)

let rec equals t1 t2 = 
    match t1, t2 with
    | Empty, Empty -> true
    | Empty, _ -> false
    | _, Empty -> false
    | Node (l1, v1, r1), Node (l2, v2, r2) -> if v1 = v2 then
                                              let eq_left = equals l1 l2 in
                                              let eq_right = equals r1 r2 in
                                              eq_left && eq_right else false ;;

(*
    Si definisca una funzione OCaML subset che prende due alberi binari t1 e t2 e 
    restituisce true se tutti i valori di t1 sono contenuti in t2, false altrimenti
*)
let rec subset t1 t2 = 
    match t1 with
    (* Empty è sottoinsieme di Node *)
    | Empty -> true 
    | Node (l, v, r) -> contains t2 v && contains t2 v ;;

(*
     Si definisca una funzione OCaML subtree che prende due alberi binari t1 e t2 
     e restituisce true se t1 è sottoalbero di t2, false altrimenti
*)

let rec is_subtree t1 t2 = 
    match t1, t2 with
    (* Se t1 è NULL è sottoalbero di t2 *)
    | Empty, Node(_,_,_) -> true
    (* Se t2 è NULL non può avere sottoalberi *)
    | Node (_,_,_), Empty -> false  
    (* t1 e t2 sono uguali o t1 è sottoalbero di t2->left o t1 è sottoalbero di t2->right *)
    | Node (_,_,_), Node (l2, v2, r2) -> equals t1 t2 || is_subtree t1 l2 || is_subtree t1 r2 ;; 
    
(*
    Si definisca una funzione OCaml che prende un albero binario e restituisce una lista contenente 
    i valori dei nodi in modo arbitrario
*)
let to_list tree = 
    let rec aux tr acc = 
        match tr with
        | Empty -> acc
        | Node (Empty, v, Empty) -> v :: acc
        | Node (l, v, r) -> [v] @ aux l acc @ aux r acc
    in aux tree [] ;;

(*
    Scrivere una funzione OCaml che presa una lista come argomento ritorni un albero binario
*)
let to_tree lst = 
    match lst with
    | [] -> Empty
    | x :: [] -> Node (Empty, x, Empty) 
    | lst' -> let rec insert tree n = 
                        match tree with
                        | Empty -> Node (Empty, n, Empty)
                        | Node (l, v, r) -> if n < v then Node (insert l n, v, r) 
                                            else if n > v then Node (l, v, insert r n) else tree in
                                            List.fold_left (fun tr x -> insert tr x) Empty lst' ;;

let t1 = Empty;;
let t2 = Node (Node (Empty, 1, Empty), 3, Empty);;
let t3 = Node (Node (Empty, 7, Node (Empty, 6, Empty)), 10, Node (Empty, 9, Empty)) ;;
let t4 = Node (Node (Node (Empty, 9, Empty), 10, Empty), 7, Empty) ;;
let t5 = Node (Node(Node (Empty, 1, Empty), 4, Node(Empty, 2, Empty)), 3, Node (Empty, 5, Empty)) ;;
let t6 = Node (Node (Empty, 1,Empty), 4,  Node (Empty, 2, Empty)) ;;

let l1 = [1; 4; 5; 6; 10] ;;
to_tree l1 ;;
```

```ocaml
 type running_sum = 
 {
     mutable sum: float;
     mutable sum_sq: float;
     mutable samples: int;
 }
 
 let mean rsum = rsum.sum /. Float.of_int rsum.samples ;;
 let stdev rsum = Float.sqrt (rsum.sum_sq /. Float.of_int rsum.samples -. mean rsum ** 2.) ;;
 
 let create () = { sum = 0.; sum_sq = 0.; samples = 0 } ;;
 
 let update rsum x = 
     rsum.samples <- rsum.samples + 1;
     rsum.sum <- rsum.sum +. x;
     rsum.sum_sq <- rsum.sum_sq +. x *. x ;;
     
let rsum = create () ;;
List.iter (fun x -> update rsum x) [1.; 3.; 2.;-7.;4.;5.];;
mean rsum ;;
stdev rsum ;;
```

```ocaml
(*
    Si definisca una funzione OCaML incBy che prende un albero binario t ed una corrispondente lista di interi lst 
    con n elementi (dove n corrisponde alla profondita di t). 
    La funzione restituisce l'albero binario ottenuto da t, 
    incrementando il valore dei nodi a livello i con il valore i-esimo nella lista.
*)

let incBy t lst = 
    let rec aux tr lvl = 
        match tr with
        | Empty -> Empty
        | Node (l, v, r) -> let incr = List.nth lst lvl in Node (aux l (lvl+1), v + incr, aux r (lvl+1)) 
    in aux t 0 ;; 
    
let tree = Node (Node (Empty, 7, Node (Empty, 6, Empty)), 10, Node (Empty, 9, Empty)) ;;
let tree2 = Node (Empty, 3, Empty) ;;
let tree3 = Empty
let list = [1; 2; 3] ;;
incBy tree list ;;
```

```ocaml
(* Marzo 2025 *)

let remove_at n lst = if List.length lst < n then failwith "Error: list's length < n" else
    let rec aux l idx = 
        match l with
        | [] -> []
        | x :: tl -> if List.nth lst x = n then tl else aux tl (idx + 1) 
    in aux lst 0 ;; 
    
let insert_at n x lst = if List.length lst < n then failwith "Error: list's length < n" else 
    let rec aux l idx = 
        match l with
        | [] -> [x]
        | hd :: tl -> if n = idx then hd :: x :: aux tl idx else aux tl (idx + 1)
    in aux lst 0 ;;
    
let apply_at n lst1 lst2 = if List.length lst1 = List.length lst2 || 
                           List.length lst1 < n + 1 && List.length lst2 < n + 1 then 
                           let rec aux l1 l2 idx =
                               match l1, l2 with
                               | [], [] -> []
                               | [], l2' -> l2'
                               | l1', [] -> []
                               | f :: f_list, hd :: tl -> if n = idx then f hd :: tl else aux f_list tl (idx+1)
                           in aux lst1 lst2 0 
                           else failwith "RunTimeError";;
        
```

```ocaml
type ntree = 
    | Leaf of string 
    | Node of string * ntree list ;;
    
let t =
  Node ("mario",
    [
      Node ("paola",
        [
          Leaf "franco";
          Node ("ada", [Leaf "raul"])
        ]
      );
      Node ("luigi",
        [
          Leaf "luca"
        ]
      );
      Node ("rosa",
        [
          Leaf "eva";
          Node ("ugo", 
            [
              Leaf "lisa";
              Leaf "lia"
            ]
          )
        ]
      )
    ]
  ) ;;
  
(* La funzione get_string restituisce la stringa corrispondente al nodo radice *)
let get_string t = match t with
    | Leaf x -> x
    | Node (x, children) -> x ;;

(* get_string_list restituisce le stringhr associate alle radice di tutti gli alberi presenti in tree_list *)
let get_string_list tree_list = List.map (fun tree -> get_string tree) tree_list ;;

(* 
    figli_di prende una stringa e un albero. Cerca il nodo o foglia con tale nome:
    
    - Nel caso lo trovi e si tratta di un nodo intermedio restituisce una lista che contiene tutti 
      i nomi associati ai nodi figli.
      
    - Altrimenti ritorna una lista vuota
*)
let rec figli_di s tree =
    let rec aux tr acc = 
        match tr with
        | Leaf x -> acc
        | Node (n, children) -> if n = s then get_string_list children @ acc
                                else List.fold_left (fun acc c -> aux c acc) [] children
    in aux tree [] ;;
    
(* 
    livello prende una stringa e un albero e restituisce il livello del nodo associato alla stringa. 
    Se non la stringa non viene trovata restituisce 0 
*)
let rec livello s t = 
    let rec aux tr lvl = 
        match tr with
        | Leaf _ -> lvl
        | Node (x, children) -> if x = s then lvl else List.fold_left (fun acc c -> aux c acc+1) 0 children 
    in aux t 0 ;;
    
livello "ada" t ;;

let rec uppercase t = 
    match t with
    | Leaf x -> Leaf (String.uppercase_ascii x) 
    | Node (s, children) -> Node (String.uppercase_ascii s, List.map uppercase children) ;;
```

```ocaml
(* Dicembre 2021 *)

type 'a m_list = 'a list ;;
type 'a mset_map = ('a * int) list ;;

(* 
    1. mult m e di tipo ‘a mset lst -> ‘a -> int che calcola la molteplicità di e nel multiset m rappre-
    sentato come lista semplice
    
    2. mult m e di tipo ‘a mset map -> ‘a -> int che calcola la molteplicità di e nel multiset m rappre-
    sentato come lista di coppie
    
    3. sum m1 m2 di tipo ‘a mset map -> ‘a mset map -> ‘a mset map 
    che calcola l’unione dei multiset m1 ed m2 rappresentati come liste di coppie 
    (esempio: sum [(’a’,2);(’b’,3)] [(’b’,4);(’c’,1)])
    restituisce [(’a’,2);(’b’,7);(’c’,1)] (eventualemnte anche con le coppie in ordine diverso)
*)

let mult m e = 
    let rec aux l acc =
        match l with
        | [] -> acc
        | x :: tl -> if x = e then aux tl (acc+1) else aux tl acc 
    in aux m 0 ;;
    
let mult_map m e = 
    let rec aux l acc =
        match l with
        | [] -> acc
        | (x, y) :: tl -> if x = e then aux tl (y + acc) else aux tl acc 
    in aux m 0 ;;
    
let sum m1 m2 = 
    let rec aux acc l1 l2 = 
        match l1, l2 with
        | [], [] -> acc
        | _, [] -> acc @ l1
        | [], _ -> acc @ l2
        | (x1, y1) :: tl1, ((x2, y2) :: _ as tl2) -> if x1 = x2 then aux ((x1, y1 + y2) :: acc) tl1 tl2
                                                     else aux acc tl1 tl2
    in aux [] m1 m2 ;;
    
let set1 = [2;4;4;3;2;3] ;;
let set2 = [2;4;3;2;2;3] ;;
let set3 = [(2,2);(4,1);(3,2)] ;;
let set4 = [(2,2);(4,1);(2,4)] ;;

sum [('a',2);('b',3)] [('b',4);('c',1)] ;;
```
